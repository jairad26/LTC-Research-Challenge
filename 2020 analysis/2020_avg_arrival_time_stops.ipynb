```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import datetime
df = pd.read_csv('data/2020 Line 280 AVL report.csv')
df['Actual Depart Hour'] = df['Actual Depart'].str.split(':',expand=True)[0]
df['Scheduled Depart Hour'] = df['Scheduled Depart'].str.split(':',expand=True)[0]

bus_actual_runtime = df.groupby(['Bus Stop ID','Actual Depart Hour'])['Actual Run Time'].mean().reset_index()
bus_expected_runtime = df.groupby(['Bus Stop ID','Scheduled Depart Hour'])['Scheduled Run Time'].mean().reset_index()


```

```python
stopactual = bus_actual_runtime.loc[bus_actual_runtime['Bus Stop ID'] == 177]
stopexpected = bus_expected_runtime.loc[bus_expected_runtime['Bus Stop ID'] == 177]
stopactual['Actual Depart Hour'] = stopactual['Actual Depart Hour'].astype(int)
stopactual = stopactual.sort_values(by=['Actual Depart Hour'])
stopactual['Actual Depart Hour'] = stopactual['Actual Depart Hour'].astype(str)


stopexpected['Scheduled Depart Hour'] = stopexpected['Scheduled Depart Hour'].astype(int)
stopexpected = stopexpected.sort_values(by=['Scheduled Depart Hour'])
stopexpected['Scheduled Depart Hour'] = stopexpected['Scheduled Depart Hour'].astype(str)

fig, ax = plt.subplots(figsize = (10,4))
plt.plot(kind="line")
maxNum = max(stopactual['Actual Run Time'].max(), stopexpected['Scheduled Run Time'].max())
ax.set(Title="2020 Bus Run Time at Stop 177 Throughout Day", xlabel="Hours of Day", ylabel="Minutes")


plt.plot(stopactual['Actual Depart Hour'], stopactual['Actual Run Time'], label = 'Actual Run Time')
plt.plot(stopexpected['Scheduled Depart Hour'], stopexpected['Scheduled Run Time'], label = 'Scheduled Run Time')

plt.grid()
plt.legend()
plt.show()
```
