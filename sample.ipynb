```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./data/2600_2020-06-01-0000_2020-06-30-0000.csv')
df.head()
```

```python
BBox = (df['GPS - GPS position Latitude'].min(),   df['GPS - GPS position Latitude'].max(),      
         df['GPS - GPS position Longitude'].min(), df['GPS - GPS position Longitude'].max())
print(BBox)
```

```python
ruh_m = plt.imread('./map.png')
```

```python
fig, ax = plt.subplots(figsize=(1,2))
ax.scatter(df['GPS - GPS position Latitude'], df['GPS - GPS position Longitude'], zorder=1, alpha=0.2, c='b',s=10)
ax.set_title('Plotting Spatial Data on Inland Empire Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```
